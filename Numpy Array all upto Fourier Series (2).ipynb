```python
fs=100
ts=np.arange(0,60,1/fs)
# initialize figure
plt.figure(figsize=[24,16])
#iterate over each amplitude and frequency and create an oscillation plot
for i in range(1,10):
    amp=100*1/float(i)
    freq=0.025*i
    
    ys=amp*np.sin(2*np.pi*freq*ts)
    
    lb="amp:"+str(amp)[0:4]+ ",freq:"+str(freq)[0:4]
    plt.plot(ts,ys, label=lb)
plt.xlabel('Time in Seconds')
plt.ylabel('Displacement')
plt.legend()
plt.show()
        
```

#### Superposition of oscillation

```python
fs=100
ts=np.arange(0,60,1/fs)
#initialize figure
plt.figure(figsize=[24,12])

'''define 10 different amplitudes and frequencies'''
amps=[10*1/float(i) for i in range(1,10)]
freqs= [0.025*i for i in range(1,10)]
'''List Comprehension'''
ys=sum([amp*np.sin(2*np.pi*freq*ts) for amp,freq in zip(amps,freqs)])
'''Create a plot'''
plt.plot(ts,ys,label='Superposed Oscillation')
plt.xlabel('Time', fontsize=22,color='magenta')
plt.ylabel('Displacement',fontsize=22,color='magenta')
plt.legend()
plt.show()
```

## Wave Propagation
A Wave function is defined as:
    $$y=Asin(\omega t + kx)$$

and the differential equation of wave is:
    $$\frac{\partial^2U}{\partial t^2}=C^2\frac{\partial^2U}{\partial x^2}$$

#### plot particle displacement keeping time fixed

```python
plt.figure(figsize=[16,2])
t0=0
fs=18
xs=np.arange(0,10,1/fs)
amp=6.0
freq=10.0
w=2*np.pi*freq
k=2*np.pi

y=amp*np.sin(w*t0-k*xs)

plt.stem(xs,y, bottom=0, label='Particle Oscillation', use_line_collection=True)
plt.xlabel('Distance', fontsize=12, color='green')
plt.ylabel('Amplitude', fontsize=12, color='green')
plt.title('Particle displacement at t0='+str(t0), fontsize=16, color='red')
plt.show()
```

```python
plt.figure(figsize=[16,18])
t0s=[0.0,0.1,0.2,0.3,0.4]
for i,t0 in enumerate(t0s):
    plt.subplot(len(t0s),1,i+1)
    y=amp*np.sin(w*t0-k*xs)
    plt.stem(y,bottom=0, label='Particle Oscillation', use_line_collection=True )
    plt.title('Particle displacement at t='+str(t0))
plt.legend()
plt.show()
    
```

## Fourier Series

#### Basis functions: {$sin(\theta),sin(2\theta),...$}

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fs=100
thetas=np.arange(0,2*np.pi,1/fs)
plt.figure(figsize=[24,12])
for i in range(1,8):
    amp=1*1/float(i)
    ys=amp*np.sin(i*thetas)
    lb="sin("+str(i)+r'$\theta$)'
    plt.plot(thetas,ys,label=lb)
plt.xlabel("Angle("+r'$\theta$)', fontsize=22,color='r')
plt.ylabel(r'$\sin(n\theta)$', fontsize=22,color='r')
plt.legend()
plt.show()
    
```

#### Basis functions: {$cos(\theta),cos(2\theta),...$}

```python
fs=100
thetas=np.arange(0,2*np.pi,1/fs)
plt.figure(figsize=[24,12])
for i in range(1,8):
    amp=1*1/float(i)
    ys=amp*np.cos(i*thetas)
    lb="cos("+str(i)+r'$\theta$)'
    plt.plot(thetas,ys,label=lb)
plt.xlabel("Angle("+r'$\theta$)', fontsize=22,color='r')
plt.ylabel(r'$\sin(n\theta)$', fontsize=22,color='r')
plt.legend()
plt.show()
```

#### Fourier Series: 
###### Square wave
$$f(x)=\frac{4}{\pi}\sum_{n=1,2,3,...}^{\infty}\frac{1}{n}\sin(\frac{n\pi x}{L})$$

##### 1. Individual basis function terms:
Here we want to plot the individual terms in the above series to compare the periodicity and amplitudes.

```python
L=1
ns=100
thetas=np.arange(0,2,1/ns)
plt.figure(figsize=[22,8])
for n in [i for i in range(10) if i%2!=0]:
    fs=(4/np.pi)*(1/n)*np.sin(n*np.pi*thetas/float(L))
    plt.plot(thetas,fs)
plt.xlabel("Angle("+r'$\theta$')
plt.ylabel("Basis Term")
plt.show()
```

##### Subsequent development by adding terms
Here we want to observe the plot of the terms(eg: first term, sum of first two terms, sum of first three terms, ...)to realize how final series is developed by the contribution of individual terms.  

```python
L=1
ns=100
thetas=np.arange(0,2,1/ns)
plt.figure(figsize=[22,8])
k=1
for n in [i for i in range(10) if i%2!=0]:
    if n==1: #taking care of first term
        
        fi=(4/np.pi)*(1/n)*np.sin(n*np.pi*thetas/float(L))
        plt.plot(thetas,fi, label='first term')
        ff=fi
    else:
        fi=ff+ (4/np.pi)*(1/n)*np.sin(n*np.pi*thetas/float(L))
        plt.plot(thetas,fi, label='sum of'+str(k)+"terms")
        ff=fi
    k=k+1
plt.xlabel("Angle("+r'$\theta$')
plt.ylabel("Sum of n Term")
plt.legend()
plt.show()
```

###### Sum of all n-terms
